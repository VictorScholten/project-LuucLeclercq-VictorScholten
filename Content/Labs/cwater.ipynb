# Bepaling van soortelijke warmte van water


# Introductie

Water heeft een enorme buffercapaciteit: je moet veel energie toevoegen om de temperatuur van water een graad te verwarmen. In dit practicum gaan we de soortelijke warmte van water bepalen door een bekende hoeveelheid water te verwarmen met een bekende hoeveelheid energie, en de temperatuurstijging te meten.

# Theorie

Om de soortelijke warmte van water te bepalen, maken we gebruik van de wet van behoud van energie. In dit experiment wordt elektrische energie omgezet in warmte, die vervolgens wordt opgenomen door het water.
Warmte en Soortelijke Warmte
Warmte ($Q$) is een vorm van energie-uitwisseling ten gevolge van een temperatuurverschil. De hoeveelheid warmte die nodig is om de temperatuur van een stof te verhogen, is afhankelijk van de massa van de stof, de gewenste temperatuurstijging en een stofeigenschap: de soortelijke warmte.
De formule hiervoor is:
$$Q = m \cdot c \cdot \Delta T$$
Waarbij:
$Q$ de toegevoegde warmte is in Joule ($\mathrm{J}$),
$m$ de massa van het water is in kilogram ($\mathrm{kg}$),
$c$ de soortelijke warmte van water is in Joule per kilogram per Kelvin ($\mathrm{J \cdot kg^{-1} \cdot K^{-1}}$) en
$\Delta T$ de temperatuurstijging is in Kelvin ($\mathrm{K}$) of graden Celsius ($^\circ \mathrm{C}$)
De soortelijke warmte van water is een bekende grootheid, doorgaans rond de $4180 \mathrm{J \cdot kg^{-1} \cdot K^{-1}}$. Dit betekent dat water relatief veel energie kan opslaan zonder sterk in temperatuur te stijgen.
De warmte in dit experiment wordt geleverd door een elektrisch verwarmingselement. De toegevoerde elektrische energie ($E_{el}$) kan worden berekend met het elektrisch vermogen ($P$) en de tijd ($t$):
$$E_{el} = P \cdot t$$
Het elektrisch vermogen ($P$) van de weerstand berekenen we met de spanning ($U$) over de weerstand en de stroomsterkte ($I$) erdoorheen:
$$P = U \cdot I$$
Als we ervan uitgaan dat er geen energieverlies optreedt naar de omgeving (zoals de lucht of de maatbeker zelf), dan is de geleverde elektrische energie gelijk aan de warmte die het water opneemt. Er geldt dan $E_{el} = Q$.

Een waterbad met bekende massa aan water wordt verwarmd met een elektrisch verwarmingselement dat een bekende hoeveelheid energie levert.
De temperatuur van het water wordt gemeten met een temperatuursensor.
Door de temperatuurstijging als functie van de tijd te meten kan de soortelijke warmte van water worden berekend.

## Procedure
```{exercise}
Meten van de massa van het water
Weeg de lege maatbeker.
Vul de maatbeker met een gewenste hoeveelheid water (200–500 ml).
Weeg de maatbeker opnieuw en bereken de massa van het water:
m_water = m(gevulde beker) – m(lege beker).
Opstellen van de meetopstelling
Plaats het verwarmingselement in het water zodat het volledig onder is, maar zonder dat het de wanden raakt.
Plaats de thermometer of temperatuursensor in het water, zodat het meetpunt volledig ondergedompeld is.
Zorg dat het water niet wordt geroerd door externe factoren; het roeren moet met een roerder, voor consistentie consistent.
Startcondities meten
Meet en noteer de begintemperatuur van het water, (T_0).
Zet de stopwatch klaar, maar start nog niet.
Water verwarmen
Stel de voedingsbron zo in dat het verwarmingselement werkt binnen zijn vermogenslimieten (10 W max).
Noteer het ingestelde vermogen P en/of de aangelegde spanning U en stroom I  (P = U * I ).
Start de voeding én de stopwatch tegelijk.
Verwarm het water gedurende een vooraf bepaalde tijd (2–5 minuten).
Meet om de 10–20 seconden de temperatuur van het water en noteer deze in een tabel.
Stoppen van de meting
Schakel de voeding na de gekozen tijd uit.
Noteer de eindtijd en de eindt temperatuut (T_e).


## Veiligheid
We maken gebruik van een $10 \mathrm{\Omega}$, $10 \mathrm{W}$ weerstand.
Deze wordt snel heet.
De bronspanning mag dan ook alleen aan wanneer de weerstand in het water zit.
Raak de weerstand niet aan tijdens het experiment.
Omdat de weerstand in het water zit, kunnen we wel het elektrisch vermogen hoger zetten zonder dat de weerstand oververhit raakt.
Het maximaal vermogen mag $40 \mathrm{W}$ zijn.
Daarbij moet de roerder wel aanstaan om de warmte goed te verdelen.

## Data analyse
Bereken de temperatuurstijging:
ΔT = T_e – T_0
Bereken de toegepaste energie:
Q = P · t
waarbij
P = vermogen van het verwarmingselement (in watt)
t = tijd dat het element aan stond (in seconden)
Bereken de soortelijke warmte (c) van water met:
c = Q/(m*ΔT)
Vergelijk het resultaat met de bekende literatuurwaarde van water:
c_water ≈ 4186  {J/(kg·°C)}

# Resultaten


````python
# Hier de data en de analyse
import numpy as np
import matplotlib.pyplot as plt

U = 12  #volt
I = 1.23 #ampepre

T = np.array([20.2,20.4,20.5, 20.7, 20.9,21.2, 21.3, 21.5, 21.6, 21.8, 22.0])
tijd =np.array([0,30,60,90,120,150,180,210, 240, 270, 300])
T_e = max(T)
T_0 = min(T)
P = U * I
delta_T =  T_e - T_0

m_water = 0.592 #kg

Q = P * max(tijd)
c_water = Q/(m_water*delta_T)
print(c_water, 'J/(K·kg)')
print('Deze gevonden waarde van de soortelijke warmte van water ligt erg dichtbij de literatuur waarde, 4186 J/(K·kg)')

plt.figure()
plt.plot(tijd, T, label="Measured water temperature")
plt.xlabel("Tijd (s)")
plt.ylabel("T (°C)")
plt.legend()
plt.savefig("figures/Measured_water_temperature.png", dpi=450)
plt.show()
````
