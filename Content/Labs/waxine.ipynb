# Waxinekaarsje

Als je een waxinekaarsje aan steekt en er een glas overheen zet, zal het kaarsje uit gaan. 
Als je dat waxinekaarsje laat drijven op water terwijl je er het glas er overheen zet, zal het water in het glas stijgen.
Op het internet zwerven er verschillende verklaringen voor dit fenomeen rond.
1. Het kaarsje verbruikt zuurstof, waardoor er een vacu√ºm ontstaat en het water omhoog wordt gezogen.
2. Het kaarsje verwarmt de lucht in het glas, waardoor de lucht uitzet. Wanneer de kaars dooft koelt de lucht af waardoor de druk afneemt en er water omhoog wordt gezogen.
3. Bij het verbranden van de kaars ontstaan er waterdamp en koolstofdioxide. De waterdamp condenseert aan de binnenkant van het glas, waardoor er minder gas in het glas is en de druk afneemt. Hierdoor wordt het water omhoog gezogen.

```{exercise}
Onderzoek welke verklaring het meest waarschijnlijk is en leg uit waarom de andere verklaringen minder waarschijnlijk zijn.
```

Om de eerste stelling te onderzoeken kan je een maatbeker gebruiken met verschillende hoeveelheden kaarsjes. Als deze stelling klopt dan maakt het niet uit hoeveel kaarsjes er in de maatbeker staat aangezien al het zuurstof wordt gebruikt ongeacht het aantal kaarsjes.
Het bleek echter dat in een maatbeker met meer kaarsjes het water uiteindelijk hoger kwam te staan dan in een zelfde maatbeker met minder kaarsjes.

Als de tweede stelling waar is moet het betekenen dat het water vooral stijgt als het kaarsje net uit is gegaan, aangezien dan het temperatuurverschil het grootst is, en neemt daarna langzaam af. Er moet ook gelden dat meer kaarsjes resulteren in een hogere uiteindelijke waterstand, omdat meerdere kaarsjes de lucht meer verwarmen.
Bij het testen bleken deze dingen beide het geval te zijn. De waterstand steeg snel toen het/de kaarsje(s) net uit waren en daarna daalde het tempo van stijgen rap. Ook resulteerde meerdere kaarsjes in een uiteindelijk hogere waterstand.

Wij namen ook condens waar bij elke test, meer kaarsjes resulteerde in meer condens. Dit condens zorgt ook voor een onderdruk aangezien het volume van water een stuk kleiner is dan de volume van gas.

Uit deze testjes hebben wij geconcludeerd dat de eerste stelling nergens op slaat, maar dat het afkoelen van de lucht en de condens allebei zorgen voor een onderdruk. Waarbij het afkoelen van de lucht meer doet aangezien er een sterke stijging van het waterpeil is in de eerste secondes na het uitgaan van de kaarsjes, maar ook daarna het water nog langzaam steigt.
